```python
from numba import jit, njit, int32
import numba
import inspect
```

```python
# @jit(nopython=False)
@njit
def add(a, b):
    res = [a + b, a, b]
    # print(numba.typeof(res))
    return res

res = add(1, 2)

print(res)

print(add.overloads.keys())

signature = add.nopython_signatures[0]
print(signature)
print(list(add.inspect_llvm().items())[0][1])
```
